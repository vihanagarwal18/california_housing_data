```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pandas.plotting import scatter_matrix

df=pd.read_csv('../input/housing/housing.csv')

count=0
for i in df['housing_median_age']:
    if i==27:
        count=count+1
print("Q1  How many blocks have a median housing age of 27?")
print("number of blocks with median housing range of 27 are:",count,sep='')
print("Q2 Add an attribute to the dataset that stores just the integer part of the latitude and determines the mean of the population for each new latitude value.")
length=len(df)
l=[]
for j in df['latitude']:
    l.append(int(j))
df['int_latitude']=l
sum_population=[]
l2=set(df['int_latitude'])
list_of_lats = list(l2)

list_of_lats = sorted(list_of_lats)

cols = ['latitude_range', 'mean_population']
out_df = pd.DataFrame(columns=cols)

for latitude in list_of_lats:
    this_lat_df = df.loc[df['int_latitude'] == latitude]
    mean = this_lat_df['population'].mean()
    records = [(latitude, mean)]
    df_new = pd.DataFrame.from_records(records, columns = cols) 

    out_df = pd.concat([out_df, df_new])
print(out_df)

print("Q3 How many blocks have the total number of bedrooms between 500 and 1000?")
bedrooms = df["total_bedrooms"]
counts_bedrooms = 0
for i in bedrooms:
    if i >= 500 and i <= 1000:
        counts_bedrooms += 1
print("No. of blocks having number of bedrooms between 500 and 1000 are: ", counts_bedrooms)



print("some graphs are:")
df.hist(bins = 10, figsize = (30,30))
plt.show()
attributes = ['median_house_value', 'median_income', "total_rooms", "housing_median_age"]

scatter_matrix(df[attributes], figsize = (12,8))

plt.show()






        

    
    
```
